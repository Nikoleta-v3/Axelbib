Status_Errors
-------------

Scraping keeps a file with the status code of each post to Axelbib. This is a very straight forward analysis
to make sure that all unique articles have been push sucessfully ones. 

Read in Report
--------------

```python
with open('/home/nikoleta/rsc/Axelbib/scraping/report', 'r') as textfile:
    report = textfile.read()
```

```python
report = report.split("\n")
print(len(report))
report = list(set(report))
len(report)
```

```python
report.remove(report[0])
```

```python
summary_report = {}
for record in report:
    key = record.split("--")[1]
    value = record.split("--")[0]
    if key in summary_report.keys() and value not in summary_report[key]:
        summary_report[key] += '--' + value
    else:
        summary_report.update({key: value})
```
